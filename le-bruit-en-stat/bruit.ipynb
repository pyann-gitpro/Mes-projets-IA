```python
import numpy as np 
import matplotlib.pyplot as plt

N = 100

# tirage uniforme
malist = list()

# N réalisation d'un processus aléatoire uniforme
for n in range(N):
    malist.append( int(np.random.rand()*6)+1 )
    
plt.figure(figsize=(8,5))
plt.plot(malist,'o')
plt.ylim((1,6))
plt.grid()
```

```python
out = list()

arr = np.array(malist)

for face in range(1,7):
    out.append(np.sum(arr==face))

print(out)

plt.bar([1,2,3,4,5,6], out)
```

```python
sigma = 1
mu = 0 

def gaussienne(x, sigma, mu):
    
    out = np.exp(- (x-mu)**2 / (2*sigma**2))
    out /= np.sqrt( 2*np.pi*sigma**2 )
    return out


x = np.linspace(-10,10,100)
g = gaussienne(x, sigma, mu)
plt.plot(x, g)

plt.plot([0, 0], [0, .5], 'k', lw=5)
g[-1]
```
